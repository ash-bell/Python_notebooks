```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import hdbscan
```

```python
df = pd.read_csv("/home/minion/projects/anvio3/bats3.umap.tsv", sep = "\t")
df.drop(["read", "length"], axis =1, inplace=True)
df.head()
```

```python
plt.figure(figsize=(18, 10), dpi= 72, facecolor='w', edgecolor='k')
clusterer = hdbscan.HDBSCAN(min_cluster_size=30, alpha=1.0).fit(df)
color_palette = sns.color_palette('bright', (clusterer.labels_.max()+1))
cluster_colors = [color_palette[x] if x >= 0
                  else (0.5, 0.5, 0.5)
                  for x in clusterer.labels_]
cluster_member_colors = [sns.desaturate(x, p) for x, p in
                         zip(cluster_colors, clusterer.probabilities_)]
plt.scatter(x = df.D1, y = df.D2, s=50, linewidth=0, c=cluster_member_colors, alpha=0.1)
#plt.savefig("clusters.png")
```
